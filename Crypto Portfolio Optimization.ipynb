# Applying Markowitz Portfolio Theory (MPT) with Efficient Frontier

```python
# Creating a table for visualising returns and volatility of assets
crypto_assets = pd.concat([crypto_means, crypto_std], axis=1) 
crypto_assets.columns = ['Returns', 'Volatility']
crypto_assets
```

```python
#the loop considers different weights for assets and calculates the return and volatility of that particular portfolio combination.

p_ret = [] # Define an empty array for portfolio returns
p_vol = [] # Define an empty array for portfolio volatility
p_weights = [] # Define an empty array for asset weights

num_assets = len(crypto_df.columns)
num_portfolios = 10000
```

```python
for portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights = weights/np.sum(weights)
    p_weights.append(weights)
    crypto_returns = np.dot(weights, crypto_means) # Returns are the product of individual expected returns of asset and its 
                                      # weights 
    p_ret.append(crypto_returns)
    var = crypto_cov_matrix.mul(weights, axis=0).mul(weights, axis=1).sum().sum()# Portfolio Variance
    crypto_std = np.sqrt(var) # Daily standard deviation
    annual_std = crypto_std*np.sqrt(365) # Annual standard deviation = volatility
    p_vol.append(annual_std)
```

```python
data = {'Returns':p_ret, 'Volatility':p_vol}

for counter, symbol in enumerate(crypto_df.columns.tolist()):
    #print(counter, symbol)
    data[symbol+' weight'] = [w[counter] for w in p_weights]
```

```python
crypto_portfolios = pd.DataFrame(data)
crypto_portfolios.head(50) # Dataframe of the 10000 portfolios created
```

```python
# Plot efficient frontier
crypto_portfolios.plot.scatter(x='Volatility', y='Returns', marker='o', s=10, alpha=0.3, grid=True, figsize=[20,20])
```

```python
#How to read the Efficient Frontier?
# Each point on the line (left edge) represents an optimal portfolio of stocks that maximises the returns for any given level of risk.
# The point (portfolios) in the interior are sub-optimal for a given risk level. 
# For every interior point, there is another that offers higher returns for the same risk.
# On this graph, you can also see the combination of weights that will give you all possible combinations:
#Minimum volatility (left most point); and Maximum returns (top most point); And everything in between.
```

```python
min_vol_port = crypto_portfolios.iloc[crypto_portfolios['Volatility'].idxmin()]
# idxmin() gives us the minimum value in the column specified.                               
min_vol_port
```

```python
# plotting the minimum volatility portfolio
plt.subplots(figsize=[30,30])
plt.scatter(crypto_portfolios['Volatility'], crypto_portfolios['Returns'],marker='o', s=10, alpha=0.3)
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=500)

#The red star denotes the most efficient portfolio with minimum volatility.
```

```python
# An optimal risky portfolio can be considered as one that has highest Sharpe ratio.
# Finding the optimal portfolio

rf = 1.564 / 100 # risk factor
optimal_risky_port = crypto_portfolios.iloc[((crypto_portfolios['Returns']-rf)/crypto_portfolios['Volatility']).idxmax()]
optimal_risky_port
```

```python
# Plotting optimal portfolio

plt.subplots(figsize=(20, 20))
plt.scatter(crypto_portfolios['Volatility'], crypto_portfolios['Returns'],marker='o', s=10, alpha=0.3)
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=500)
plt.scatter(optimal_risky_port[1], optimal_risky_port[0], color='g', marker='*', s=500)

#The green star represents the optimal risky portfolio.
```
